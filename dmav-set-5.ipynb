```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[10,20,30,40,50]
```

```python
plt.plot(x,y,marker='o')
plt.xlabel("X values (independent)")
plt.ylabel("Y values (dependent)")
plt.title("X-Y Axis Data plot")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

n = int(input("Enter number of data points: "))

x = []
y = []

for i in range(n):
    xi = float(input(f"Enter x{i+1}: "))
    yi = float(input(f"Enter y{i+1}: "))
    x.append(xi)
    y.append(yi)

plt.plot(x, y, marker="o")

plt.xlabel("X - ind")
plt.ylabel("Y - d")
plt.title("X-Y Axis Data")

plt.grid(True)
plt.show()
```
